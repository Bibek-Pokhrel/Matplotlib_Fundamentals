```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.plot(x,y)       # builds the linear graph
plt.show()
```

```python
# without linear value
x=[1,5,3,2,5]
y=[10,5,9,7,6]
c='r'               #gives red color to graph
plt.plot(x,y,c,marker="o")  #marker gives the dot in the edge
plt.show()
```

```python
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.bar(x,y)        #to plot bargraph
```

```python
#to change the color of bargraph
x=[1,2,3,4,5]
y=[4,5,6,7,8]
c=['r','g','b','y']
plt.bar(x,y,color=c)    # plt.barh(x,y) gives horizontal graph
```

```python
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.bar(x,y,width=0.3)  #width gives the width of the graph
```

```python
x=[1,2,3,4,5]
y=[4,5,6,7,8]
plt.barh(x,y,height=0.1)    #barh gives the horizontal bar graph
```

```python
x=[1,2,3,4,5,4,3,7]
y=[4,5,6,7,8,6,3,4]
plt.title("scratter plot")      #gives the heading to the plot
plt.scatter(x,y)                #gives the dot on the coordinate
```

```python
# to give label to x and y axis
x=[1,2,3,4,5,6,7]
y=[10000,16000,9000,14000,15000,21000,19000]
plt.xlabel("months")    #gives label to x axis
plt.ylabel("sales")     #gives label to y axis
plt.title("sales report")      #gives the heading
plt.plot(x,y)
```

```python
# for Pie chart
x=['jan','feb','mar','apr','may','jun','july']
y=[10000,16000,9000,14000,15000,21000,19000]
plt.title("sales chart")
plt.pie(y,labels=x)
plt.legend()
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
fname = r'image.png'    #read image in readmode
image=Image.open(fname) #open image
plt.imshow(image)
```

```python
image=Image.open(fname).convert('L')    #convert(L) convert image into gray scale L stands for luminance
plt.imshow(image,cmap="gray")           # cmap converts the image into give color
plt.show()
```

```python
image=Image.open(fname).convert('L')
plt.imshow(image,cmap="coolwarm")       # here cmap changes the image into coolwarm
```
